# Python Branching and Loops


## * Branching


Branching allows the program to take decisions based on conditions.
It provides:
- if
- elif
- else


### Syntax


```python
if condition:
    statement             #runs when condition is true
elif condition:
    statement             #runs when the aabove if is false and this is true
else:
    statement            #runs when all conditons above are false
```

### Example


```python
age = 20

if age >= 18:
    print("Eligible to vote")
else:
    print("Not eligible")

#example

marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
else:
    print("Grade C")
```

## * Loops


Loops are used to repeat a block of code until a condition is met.
The body of a loop is separated using indentation.



Types of loops:
- for loop
- while loop


## * for Loop


Used when we already know how many times to iterate. 

### Syntax


```python
for variable in sequence:
    statement     #repeated code
```

### Example


```python
for i in [1, 2, 3, 4]:
    print(i)
```

## * range()

range() is used to generate a sequence of numbers.


### Types of range:
- range(stop)
- range(start, stop)
- range(start, stop, step)


### Examples


```python
list(range(6))
```

```python
list(range(1, 6))
```

```python
list(range(0, 13, 1))
```

```python
list(range(2, 25, 5)) #step means jumping
```

## * for Loop with range


```python
for i in range(1, 6):
    print(i)
```

## * while Loop


The while loop runs as long as the condition is True.
It is used when the number of iterations is not known in advance.


### Syntax


```python
while condition:
    statement
```

### Example


```python
i = 1

while i <= 5:
    print(i)
    i = i + 1
```

## * break Statement


Used to stop the loop immediately.


```python
for i in range(1, 10):
    if i == 5:
        break
    print(i)
```

## * continue Statement


Used to skip the current iteration.


```python
for i in range(1, 6):
    if i == 3:
        continue      #3 is skipped
    print(i)
```

## * Summary


- if, elif, else are used for branching
- for loop is used when iteration count is known
- while loop is used when iteration count is unknown
- range() generates numbers
- break stops the loop
- continue skips the iteration
