# String Slicing
Use colon notation to slice strings.


```python
s[0:]      # from start
s[:5]      # up to index 5 (not including)
s[:]       # entire string
```

# Lists
Lists are mutable sequences.


```python
my_list = [1, 2, 3]
my_list = ["A string", 23, 100.232, "o"]
len(my_list)
```

# Indexing and Slicing Lists


```python
my_list = ["one", "two", "three", 4, 5]

my_list[3]
my_list[1:]
my_list[:5:2]
```

# List Concatenation and Reassignment


```python
my_list + ["new item", 5]
my_list = my_list + ["add new item permanently"]
my_list
```

# List Multiplication


```python
my_list * 5
```

# Basic List Methods


```python
l = [1,2,3]
l.append("append me!")
l
```

```python
l = [1,2,3]
l.pop()
l
```

```python
new_list = ['a','e','x','b','c']
new_list.reverse()
new_list.sort()
new_list
```

# Nesting Lists


```python
lst_1 = [1,2,3]
lst_2 = [4,5,6]
lst_3 = [7,8,9]

matrix = [lst_1, lst_2, lst_3]
matrix
matrix[1][1]
matrix[0]
matrix[0][0]
```

# List Comprehensions


```python
first_col = [row[0] for row in matrix]
first_col
```

```python
first_col = []
for row in matrix:
    first_col.append(row[0])
first_col
```
